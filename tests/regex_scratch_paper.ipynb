```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import re
from darkplot import *
```

```python
with open('example data/Experiment 2025-1-24 SA 450.xy', 'r') as f:
    text = f.read()
new_entry_pattern = r"\n\n"
entries = re.split(new_entry_pattern, text)
```

```python
operations_pattern = r"# Operation:"
operations = []

spectrum_pattern = r"# Region:"
spectra = []
```

```python
for entry in entries:
    if any(re.findall(operations_pattern, entry)):
        operations.append(entry)
    if any(re.findall(spectrum_pattern, entry)):
        spectra.append(entry)
```
